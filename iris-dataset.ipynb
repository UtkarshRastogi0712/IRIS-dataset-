```python
dataset= pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
data=dataset.to_numpy()
np.random.shuffle(data)
training_set=np.array(data[:120,:4], dtype=np.float32)
test_set=np.array(data[120:,:4], dtype=np.float32)

x_train=((training_set-(np.min(training_set, axis=0)))/np.ptp(training_set, axis=0))
x_test=((test_set-(np.min(test_set, axis=0)))/np.ptp(test_set, axis=0))

labels=data[:,4]
unique, inverse=np.unique(data[:,4], return_inverse=True)
onehot = np.eye(unique.shape[0])[inverse]
y_train=tf.convert_to_tensor(onehot[:120,:])
y_test=tf.convert_to_tensor(onehot[120:,:])
```

```python
model= models.Sequential()
model.add(layers.Dense(4, activation='relu'))
model.add(layers.Dense(5, activation='relu'))
model.add(layers.Dense(5, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))
model.add(layers.Dense(3))
optimizer1 = optimizers.Adam(learning_rate=0.001)
optimizer2 = optimizers.Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer1,
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history1 = model.fit(np.array(x_train), np.array(y_train), epochs=500, 
                    validation_data=(np.array(x_test), np.array(y_test)))
```

```python
model.compile(optimizer=optimizer2,
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
history2 = model.fit(np.array(x_train), np.array(y_train), epochs=500, 
                    validation_data=(np.array(x_test), np.array(y_test)))
```

```python
plt.plot(history1.history['val_accuracy'], label='accuracy')
plt.plot(history1.history['val_loss'], label='loss')
plt.plot(history2.history['val_accuracy'], label='accuracy')
plt.plot(history2.history['val_loss'], label='loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```
